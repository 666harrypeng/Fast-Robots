```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft
```

```python
# read sample data
data = []

with open('gyr_data_sample.txt', 'r') as file:
    for line in file:
        # Split the line into values and convert them to the appropriate data type
        values = line.strip().split(',')
        data.append(values)

data = np.array(data).astype(float)
print(np.shape(data))
```

```python
x = np.arange(0, np.shape(data)[0])
plt.plot(x, data[:, 0], label="pitch", color="red")
plt.plot(x, data[:, 1], label="roll", color="blue")
plt.plot(x, data[:, 2], label="yaw", color="green")
plt.ylabel("degree")
plt.xlabel("sample count")
plt.legend()
plt.show()
```
